```python
import pandas as pd
from datetime import datetime
import matplotlib.pyplot as plt
```

```python
plt.style.use('dark_background')
```

```python
df_sm = pd.read_csv('./data/nakamura_1979_sm_locations.csv')
df_sm['Time'] = [datetime.strptime(i, '%Y%j').date() for i in df_sm['Year'].astype(str)+''+df_sm['Day'].astype(str)]
df_sm['Date'] = pd.to_datetime(df_sm['Time'].astype(str)+' '+df_sm['H'].astype(str)+':'+df_sm['M'].astype(str)+':'+df_sm['S'].astype(str))
fig, ax = plt.subplots(figsize=(8,3),dpi=200)
ax.plot(df_sm['Date'],df_sm['Magnitude'], marker='o', mec='brown', ls='-.', c='yellow')
plt.title('Shallow moonquake - Nakamura (2005)')
plt.ylabel('Magnitude')
plt.xlabel('Year')
fig.autofmt_xdate()
plt.margins(x=0.009)
plt.tight_layout()
ax.grid(False)
plt.savefig('Shallow_moonquake.png', dpi=200)
```

```python
#plotting septh
df_dm = pd.read_csv('./data/nakamura_2005_dm_locations.csv')
fig, ax2 = plt.subplots(figsize=(8,3),dpi=200)
ax2.plot(df_dm.index, df_dm['Depth'] , marker='o', mec='brown', c='yellow', ls='-.', lw=0.8, ms=5)
plt.xlabel('Seismic event')
plt.ylabel('Depth [km]')
plt.fill_between(x=df_dm.index,
                 y1=df_dm['Depth'] - df_dm['Depth_err'],
                 y2=df_dm['Depth'] + df_dm['Depth_err'],
                 alpha=0.25, color='white'
                 )
plt.tight_layout()
plt.savefig('nakamura_dm_locations.png', dpi=150)
```
