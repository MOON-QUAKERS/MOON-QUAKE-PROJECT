```python
import pandas as pd
```

```python
files = [ "nakamura_1983_ai_locations.csv", "nakamura_1979_sm_locations.csv", "nakamura_2005_dm_locations.csv"]
df_tot = pd.DataFrame()
f = False

cols = ["Lat", "Long"]
dfa = []

#Grab lat/long coords
for file in files:
    df = pd.read_csv( file )
    for index,data in df.iterrows():
        dfa.append( {"Lat":data.Lat, "Long":data.Long} )
    
df = pd.DataFrame( dfa )
df.head()
```

```python
from PIL import Image, ImageDraw

#Static image size
x, y = 16384, 8192
cx = int( x/2 )
cy = int( y/2 )
degX = cx/180
degY = cy/180

#Convert lat/long to XY
# dummy estimation function
def latlngToGlobalXY(lat, lng):
    x = cx + (degX * lng)
    y = cy + (degY * lat)
    return x, y

# Generate overlay image / Mask of areas with seismic activity
image   = Image.new( size=(x,y), mode="1", color="black")
draw    = ImageDraw.Draw(image)
for a in dfa:
    tx,ty = latlngToGlobalXY( a["Lat"], a["Long"] )
    print( f'[{a["Long"]}, {a["Lat"]}],')
    draw.ellipse([(tx-15,ty-15),(tx+15,ty+15)], fill="white")

image.save("seismic_overlay.jpg") 
```
